```python
df = df.drop_duplicates().reset_index(drop=True)
```

```python
df = df[['title', 'overview', 'genres','tagline','vote_average','popularity']]
```

```python
df.head()
```

```python
df = df.dropna(subset=['title']).reset_index(drop=True)
```

```python
df['overview'] = df['overview'].fillna('')
```

```python
df.iloc[0]['genres']
```

```python
import ast
df['genres'] = df['genres'].apply(lambda x: [i['name'] for i in ast.literal_eval(x)])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['tagline'] = df['tagline'].fillna('')
```

```python
df['tags'] = df['overview'] + " " + df['genres'].apply(lambda x: " ".join(x)) + " " + df['tagline']
```

```python
df['tags'][1]
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess(text):
    text = text.lower()
    text = re.sub('[^a-zA-Z]', ' ', text)
    words = text.split()
    words = [lemmatizer.lemmatize(word) for word in words if word not in stopwords.words('english')]

    return " ".join(words)
```

```python
df['tags'] = df['tags'].apply(preprocess)
```

```python
df = df.reset_index(drop = True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix
```

```python
indices = pd.Series(df.index,index = df['title']).drop_duplicates()
indices
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title, n = 10):
  if title not in indices:
    return ['Movie not found']

  idx = indices[title]
  sim_score = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
  similar_idx = sim_score.argsort()[::-1][1:n+1]
  return df['title'].iloc[similar_idx]
```

```python
recommend('Avatar',5)
```

```python
import pickle

pickle.dump(tfidf_matrix,open('tfidf_matrix.pkl','wb'))

pickle.dump(indices,open('indices.pkl','wb'))

df.to_pickle('df.pkl')

pickle.dump(tfidf,open('tfidf.pkl','wb'))
```
